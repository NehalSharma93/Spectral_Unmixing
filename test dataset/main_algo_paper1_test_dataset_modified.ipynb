```python
import numpy as np
from scipy.optimize import minimize
from osgeo import gdal
import sys
import math
# this allows GDAL to throw Python Exceptions
gdal.UseExceptions()

def objective(x):
    #print("inside objective")
    squaredErrorSum = (np.square(r[0] - (a[0,0]*x[0] + a[0,1]*x[1] + a[0,2]*x[2])) + \
                    np.square(r[1] - (a[1,0]*x[0] + a[1,1]*x[1] + a[1,2]*x[2])) + \
                    np.square(r[2] - (a[2,0]*x[0] + a[2,1]*x[1] + a[2,2]*x[2])) + \
                    np.square(r[3] - (a[3,0]*x[0] + a[3,1]*x[1] + a[3,2]*x[2])) + \
                    np.square(r[4] - (a[4,0]*x[0] + a[4,1]*x[1] + a[4,2]*x[2])) + \
                    np.square(r[5] - (a[5,0]*x[0] + a[5,1]*x[1] + a[5,2]*x[2])))
#     print(squaredErrorSum)
    return squaredErrorSum

def constraint1(x):
    return x[0]+x[1]+x[2]-1.0



band1 = gdal.Open("geotiff_testdata_band1")
band2 = gdal.Open("geotiff_testdata_band2")
band3 = gdal.Open("geotiff_testdata_band3")
band4 = gdal.Open("geotiff_testdata_band4")
band5 = gdal.Open("geotiff_testdata_band5")
band6 = gdal.Open("geotiff_testdata_band6")

abun1 = gdal.Open("geotiff_testdata_abundance_class1")
abun2 = gdal.Open("geotiff_testdata_abundance_class2")
abun3 = gdal.Open("geotiff_testdata_abundance_class3")

srcabun1 = abun1.GetRasterBand(1)
srcabun2 = abun1.GetRasterBand(1)
srcabun3 = abun1.GetRasterBand(1)

srcband1 = band1.GetRasterBand(1)
srcband2 = band2.GetRasterBand(1)
srcband3 = band3.GetRasterBand(1)
srcband4 = band4.GetRasterBand(1)
srcband5 = band5.GetRasterBand(1)
srcband6 = band6.GetRasterBand(1)
    
class1_abundance = srcabun1.ReadAsArray()
class2_abundance = srcabun2.ReadAsArray()
class3_abundance = srcabun3.ReadAsArray()

# class1_abundance = class1_abundance.ravel()
# class2_abundance = class2_abundance.ravel()
# class3_abundance = class3_abundance.ravel()

srcband1_array = srcband1.ReadAsArray()
srcband2_array = srcband2.ReadAsArray()
srcband3_array = srcband3.ReadAsArray()
srcband4_array = srcband4.ReadAsArray()
srcband5_array = srcband5.ReadAsArray()
srcband6_array = srcband6.ReadAsArray()

result_class1 = list()
result_class2 = list()
result_class3 = list()

# spectral library provided by Uttam sir - Substrate, Vegetation, Dark Matter
a = np.matrix([[1781.88,301.22,188.75], 
            [3371.55,607,104.53],
            [4580.77,312.22,49.21],
            [5586,6687.88,70.65],
            [6828.22,2407.88,29.61],
            [6447.44,958.11,22.24]])
# print(a.shape)


# r_list = [[srcband1_array[11][0]],
#           [srcband2_array[11][0]],
#           [srcband3_array[11][0]],
#           [srcband4_array[11][0]],
#           [srcband5_array[11][0]],
#           [srcband6_array[11][0]]]



# r_list = [[srcband1_array[0][0]],
#           [srcband2_array[0][0]],
#           [srcband3_array[0][0]],
#           [srcband4_array[0][0]],
#           [srcband5_array[0][0]],
#           [srcband6_array[0][0]]]

# r_list = [[srcband1_array[1118][4]],
#           [srcband2_array[1118][4]],
#           [srcband3_array[1118][4]],
#           [srcband4_array[1118][4]],
#           [srcband5_array[1118][4]],
#           [srcband6_array[1118][4]]]



# for i in range(100, 110):
#     for j in range(100, 110):
# for i in range(300, 310):
#     for j in range(300, 310):
# for i in range(500, 510):
#     for j in range(500, 510):
for i in range(0,10):
    for j in range(0,10):
        r_list = [[srcband1_array[i][j]],
                  [srcband2_array[i][j]],
                  [srcband3_array[i][j]],
                  [srcband4_array[i][j]],
                  [srcband5_array[i][j]],
                  [srcband6_array[i][j]]]

        #print(r_list)

        r = np.array(r_list)
        #     print(r)
        #     print(r.shape)

        n = 3
        x0 = np.zeros(n)
#         initial guess values
        x0[0] = 0.33
        x0[1] = 0.33
        x0[2] = 0.33

        print("for pixel: " + str(i) + "," + str(j))
        # show initial objective
        print('Initial Objective: ' + str(objective(x0)))

        # optimize
        b = (0.0,1.0)
        bnds = (b, b, b)

        con1 = {'type': 'eq', 'fun': constraint1} 
        cons = ([con1])

        #SLSQP=Sequential Least SQuares Programming optimization algorithm 

        sol = minimize(objective,x0,method='SLSQP',bounds=bnds,constraints=cons,options={'eps': 1.4901161193847656e-01, 'maxiter' : 50})
        x = sol.x
#         print(x)
        # show final objective
        print('Final Objective: ' + str(objective(x)))

        # print solution
#         print('Solution')
        
#         print('x1 = ' + str(x[0]))
#         print('x2 = ' + str(x[1]))
#         print('x3 = ' + str(x[2]))
#         print('sum = ', x[0] + x[1] + x[2])
#         print("-----------------------------")
        
#         result_class1.append(x[0])
#         result_class2.append(x[1])
#         result_class3.append(x[2])
        if(x[0] == 0.33 and x[1] == 0.33 and x[2] == 0.33):
            result_class1.append(-1000)
            result_class2.append(-1000)
            result_class3.append(-1000)
        else:
            if(math.isnan(x[0])):
                x[0] = -1000
            result_class1.append(x[0])
            if(math.isnan(x[1])):
                x[1] = -1000
            result_class2.append(x[1])
            if(math.isnan(x[2])):
                x[2] = -1000
            result_class3.append(x[2])

class1_abundance = np.array(class1_abundance)
result_class1 = np.array(result_class1)
count=0 
totcount = 0
for i in range(0, result_class1.shape[0]):
    if(result_class1[i] == -1000):
        count = count + 1
    totcount = totcount + 1
print(count)
print(totcount)

class2_abundance = np.array(class2_abundance)
result_class2 = np.array(result_class2)
count=0 
totcount = 0
for i in range(0, result_class2.shape[0]):
    if(result_class2[i] == -1000):
        count = count + 1
    totcount = totcount + 1
print(count)
print(totcount)

class3_abundance = np.array(class3_abundance)
result_class3 = np.array(result_class3)
count=0 
totcount = 0
for i in range(0, result_class3.shape[0]):
    if(result_class3[i] == -1000):
        count = count + 1
    totcount = totcount + 1
print(count)
print(totcount)

# print(class1_abundance)
# print(result_class1)

## Logic for comparison of the abundance_observed_result with the abundance_estimated_result

# print(class1_abundance)
# print(result_class1)
diff_list_class1 = list()
diff_list_class2 = list()
diff_list_class3 = list()

class1_abundance = class1_abundance.ravel()
class2_abundance = class2_abundance.ravel()
class3_abundance = class3_abundance.ravel()


for i in range(0,(10*10)):
    if(result_class1[i] == -1000):
        x=0
    else:
        x = result_class1[i] - class1_abundance[i]
    diff_list_class1.append(x)
    
    if(result_class2[i] == -1000):
        x=0
    else:
        x = result_class2[i] - class2_abundance[i]
    diff_list_class2.append(x)
    
    if(result_class3[i] == -1000):
        x=0
    else:
        x = result_class3[i] - class3_abundance[i]
    diff_list_class3.append(x)


diff_list_array1 = np.array(diff_list_class1)
print(diff_list_array1.shape[0])

diff_list_array2 = np.array(diff_list_class2)
print(diff_list_array2.shape[0])

diff_list_array3 = np.array(diff_list_class3)
print(diff_list_array3.shape[0])



rmse_class1 = 0
for i in range(0, diff_list_array1.shape[0]):
    rmse_class1 = rmse_class1 + np.square(diff_list_array1[i])
    
rmse_class2 = 0
for i in range(0, diff_list_array2.shape[0]):
    rmse_class2 = rmse_class2 + np.square(diff_list_array2[i])

rmse_class3 = 0
for i in range(0, diff_list_array3.shape[0]):
    rmse_class3 = rmse_class3 + np.square(diff_list_array3[i])

print(rmse_class1)
print(rmse_class2)
print(rmse_class3)

rmse_class1 = rmse_class1/diff_list_array1.shape[0]
rmse_class2 = rmse_class2/diff_list_array2.shape[0]
rmse_class3 = rmse_class3/diff_list_array3.shape[0]

print(rmse_class1)
print(rmse_class2)
print(rmse_class3)

    
        
```
